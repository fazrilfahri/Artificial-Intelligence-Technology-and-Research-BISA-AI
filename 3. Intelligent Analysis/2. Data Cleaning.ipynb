# Cleaning Data

## Produk Dibatalkan - obat tradisional

```python
df = pd.read_csv('/content/drive/MyDrive/SIB/Project/Produk Dibatalkan/obat tradisional.csv')
df.head()
```

```python
# lower casing
cols = df.columns[1:]
for col in cols:
    df[col] = df[col].str.lower()

# feature extraction
for i in df['status']:
    x = i.split()[:2]
    y = i.split()[-1]
    stats = ' '.join([x for x in x])
    alasan = ' '.join([y.replace('(', '').replace(')', '')])
df['status'] = stats
df['alasan'] = alasan

# drop duplicated
df = df.drop_duplicates(subset = ['registration number'])

df.to_csv('/content/drive/MyDrive/SIB/Project/Produk Dibatalkan/clean - obat tradisional.csv', index = False)
```

## Produk Dibatalkan - Produk Pangan

```python
df = pd.read_csv('/content/drive/MyDrive/SIB/Project/Produk Dibatalkan/produk pangan.csv')
df.head()
```

```python
# lower casing
cols = df.columns[1:]
for col in cols:
    df[col] = df[col].str.lower()

# feature extraction
stats = []
alasan = []
for i in df['status']:
    x = i.split('(')[0]
    x = x.strip()
    y = i.split('(')[1]
    stats.append(x)
    alasan.append(y.replace(')', ''))
    # alasan = ' '.join([y.replace(')', '')])
df['status'] = stats
df['alasan'] = alasan

# drop duplicated
df = df.drop_duplicates(subset = ['registration number'])

df.to_csv('/content/drive/MyDrive/SIB/Project/Produk Dibatalkan/clean - produk pangan.csv', index = False)
```
