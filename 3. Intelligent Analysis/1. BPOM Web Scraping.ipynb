# Import Library

```python
import pandas as pd
```

# Function for Cleaning Data

```python
def clean_df(regnum, prod_name, stat, prod, file_name):
    #### registration number
    
    # sort values
    idx = []
    for i in regnum[regnum.columns[0]].values:
        i = i.split('-')[1]
        i = int(i)
        idx.append(i)
    regnum[regnum.columns[0]] = idx
    regnum = regnum.sort_values(regnum.columns[0], ascending = True, ignore_index = True)
    
    # clean regnum
    registration_number = []
    for reg in list(regnum[regnum.columns[-1]]):
        reg = reg.split(sep = 'Terbit')[0]
        registration_number.append(reg)
    regnum[regnum.columns[-1]] = registration_number
    
    ### product name
    
    # sort values
    idx = []
    for i in prod_name[prod_name.columns[0]].values:
        i = i.split('-')[1]
        i = int(i)
        idx.append(i)
    prod_name[prod_name.columns[0]] = idx
    prod_name = prod_name.sort_values(prod_name.columns[0], ascending = True, ignore_index = True)
    
    # clean prod_name
    product_name = []
    for product in list(prod_name[prod_name.columns[-1]]):
        product = product.split(sep = 'Merk')[0]
        product_name.append(product)
    prod_name[prod_name.columns[-1]] = product_name
    
    
    # product
    prod = [prod] * len(regnum)
    prod_df = pd.DataFrame(prod)
    prod_df.columns = ['product']
    
    ### status
    
    # sort values
    idx = []
    for i in stat[stat.columns[0]].values:
        i = i.split('-')[1]
        i = int(i)
        idx.append(i)
    stat[stat.columns[0]] = idx
    stat = stat.sort_values(stat.columns[0], ascending = True, ignore_index = True)
    
    status = stat[stat.columns[-1]]
    
    ### concat
    df_clean = pd.concat([regnum[regnum.columns[-1]], prod_name[prod_name.columns[-1]],
                          prod_df, status], axis = 1)
    df_clean.columns = ['registration number', 'product name', 'product', 'status']
    
    ### export to csv
    df_clean.to_csv(file_name, index = False)
    
    return df_clean
```

# Produk Obat Tradisional Ditarik

```python
reg_num = pd.read_csv('Obat Tradisional/tradisional - regnum 1.csv')
products = pd.read_csv('Obat Tradisional/tradisional - produk 1.csv')
status = pd.read_csv('Obat Tradisional/tradisional - status 1.csv')
prod = 'Obat Tradisional'
file_name = 'obat tradisional.csv'
clean_df(reg_num, products, status, prod, file_name).head()
```
